```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x+1,color='red',linewidth=0.25)
ax.plot(x,x+2,color='red',linewidth=0.50)
ax.plot(x,x+3,color='red',linewidth=1.00)
ax.plot(x,x+4,color='red',linewidth=2.00)
# possible linestype options -, , -., :, steps
ax.plot(x,x+5,color='blue',lw=3,linestyle='-')
ax.plot(x,x+6,color='blue',lw=3,linestyle='-.')
ax.plot(x,x+7,color='blue',lw=3,linestyle=':')
# custom dash
line, =ax.plot(x,x+8,color='purple',lw=1.50)
line.set_dashes([5,10,15,10])
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x,x+9,color='black',lw=3,ls='-',marker='+')
ax.plot(x,x+10,color='black',lw=3,ls='-.',marker='o')
ax.plot(x,x+11,color='black',lw=3,ls=':',marker='s')
ax.plot(x,x+12,color='black',lw=3,ls=':',marker='1')
# marker size and color
ax.plot(x,x+13,color='green',lw=1,ls='-',marker='o',markersize=2)
ax.plot(x,x+14,color='green',lw=1,ls='-',marker='o',markersize=4)
ax.plot(x,x+15,color='green',lw=1,ls='-',marker='o',markersize=8,markerfacecolor='red')
ax.plot(x,x+16,color='green',lw=1,ls='-',marker='s',markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");

                    
                    
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")
axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")
axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
